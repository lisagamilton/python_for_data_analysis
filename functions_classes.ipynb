```python
#add 10 to each number
nums = [1, 5, 10]
num1 = [6, 3, 11]

def add10(numbers):
  return [n+10 for n in numbers]

add10(nums)
```

```python
#turn to string
nums = [1, 2, 3]

def to_string(numbers):
  return [str(n) for n in numbers]

to_string(nums)
```

```python
#only big numbers (10) and multiply by 2
nums = [5, 20, 3, 15]

def calculate(numbers):
  return [n*2 for n in numbers if n > 10]

calculate(nums)
```

```python
#return length
words = ["data", "python", "ai"]
words1 = ['dfksg','maybe','God bless America']
def length(words):
  return [len(word) for word in words]

length(words)
length(words1)
```

```python
#clean text
names = [" Alice ", "BOB", "  Charlie "]

def clean_text(text):
  return [t.strip().capitalize() for t in text]
clean_text(names)
```

```python
#10% off of each price
prices = [100, 200, 300]

def ten_percent(prices):
  return [p * 0.9 for p in prices]

ten_percent(prices)
```

```python
#only products more than 100
prices = [50, 120, 80, 200]

def filter(prices):
  return [p for p in prices if p > 100]
filter(prices)
```

```python
#double salary
salaries = [1000, 1500, 2000]
def double(salaries):
  return [s * 2 for s in salaries]

double(salaries)
```

```python
#email clean
emails = ["  TEST@GMAIL.COM ", "USER@MAIL.COM  "]
def clean_email(emails):
  return [email.strip().lower() for email in emails]

clean_email(emails)
```

```python
#count true
logins = [True, False, True, True, False]
def count(logins):
  return sum(logins)

count(logins)
```

```python
#filter only successful orders
orders = ["success", "failed", "success", "pending"]
def filter(orders):
  return [o for o in orders if o == 'success']
filter(orders)
```

```python
#Add VAT price 20%
prices = [100, 200, 300]

def vat(prices):
  return [p + p * 0.2 for p in prices]

vat(prices)
```

```python
#length name of products
products = ["laptop", "phone", "tv"]

def length(products):
  return [len(p) for p in products]
length(products)
```

```python
#Data cleaning (None)
revenues = [1000, None, 2000, None, 3000]
def clean(data):
  return [d for d in data if d != None]

clean(revenues)
```

```python
#check if there is a big check (>100)
orders = [50, 80, 120, 40]
def check(orders):
  return any([o for o in orders if o > 100])

check(orders)
```

# Classes

```python
class ItSpecialist:
  def __init__(self, name, speciality, salary):
    self.name = name
    self.speciality = speciality
    self.salary = salary

  def tell_about_yourself(self):
    return f'Hi! My name is {self.name}. I work in {self.speciality} field and earn {self.salary}'
#name is not defined here so we refer to self

john = ItSpecialist('John','DS',15000)
john.tell_about_yourself()
```

```python
class Student:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def introduce(self):
    return f'Hi, my name is {self.name} and I am {self.age} years old'

  def is_adult(self):
    if self.age >= 18:
      return 'Adult'
    else:
      return 'Minor'

student1 = Student('John',20)
student1.introduce()
student1.is_adult()
```
