```python
#if else
age = 20
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
# even or odd
num = 7

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#while
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
#even numbers
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = [n for n in numbers if n % 2 == 0]
print(even_numbers)
```

```python
#square numbers
squares = [n**2 for n in numbers]
print(squares)
```

```python
#count apples
items = ['apple', 'banana', 'apple', 'cherry', 'apple']

count = 0
for item in items:
    if item == 'apple':
        count += 1
print("Apples:", count)
```

```python
#removing duplicates
items = ['a', 'b', 'a', 'c', 'b']
unique = []
for item in items:
    if item not in unique:
        unique.append(item)
print(unique)
```

```python
#string cleaning
text = "  Hello World  "
cleaned = text.strip().lower()
print(cleaned)
```

```python
#sum and avg
numbers = [10, 20, 30, 40]
total = 0
for n in numbers:
    total += n
average = total / len(numbers)

print("Sum:", total)
print("Average:", average)
```

```python
#temp filter
temps = [12, 25, 8, 30, 15]
result = [t for t in temps if t>20]
print(result)
```

```python
#count positive numbers
nums = [-3, 10, -1, 5, 0, 7]
count = sum([1 for num in nums if num > 0])
print(count)
```

```python
#replace if <50 then 50
scores = [40, 55, 70, 30]
new_score = [50 if score < 50 else score for score in scores]
print(new_score)
```

```python
#shortest word
words = ["analytics", "data", "python", "ai"]
shortest = min(words,key = len)
print(shortest)
```

```python
#delete negative numbers
nums = [-5, 3, -2, 7, 1]
cleaned = [n for n in nums if n >= 0]
print(cleaned)
```

```python
#check if there is at least one big number (bigger than 50)
nums = [10, 20, 30, 100]
result = any(n > 50 for n in nums)
print(result)
```

```python
#make all words capitalized
names = ["anna", "john", "mike"]
capital_names = [name.upper() for name in names]
print(capital_names)
```

```python
#count odd and even
nums = [1, 2, 3, 4, 5, 6]
even = sum([1 for n in nums if n % 2 == 0])
odd = sum([1 for n in nums if n%2 !=0])
print(even)
print(odd)
```
