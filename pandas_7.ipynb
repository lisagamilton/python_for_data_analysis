```python
# Cleaning data
data_list = ['my',33,1.73,'test',4.6,33, 'test']
```

```python
numbers = []
strings = []
```

```python
for item in data_list:
  if isinstance(item, (int,float)):
    if item not in numbers: #making sure there are no duplicates
      numbers.append(item)
      print(numbers)
  elif isinstance(item,str):
    if item not in strings:
      strings.append(item)
      print(item)
```

```python
print(numbers)
print(strings)
```

```python
data_list = [
    ['apple',3,1.5,'banana'],
    [7, 'apple', 4.6, 1.5],
    ['banana',2,7,'cherry']
]
data_list
```

```python
unique_numbers = []
unique_strings = []

for sublist in data_list:
  for item in sublist:
    if isinstance(item,(int, float)):
      if item not in unique_numbers:
        unique_numbers.append(item)
        print(item)
    elif isinstance(item, str):
      if item not in unique_strings:
        unique_strings.append(item)
        print(item)
```

```python
print(unique_numbers)
print(unique_strings)
```

```python
#What if column names are different for join?
clients_df = pd.DataFrame({
    'ClientID':[1,2,3,4],
    'ClientName':['Alice','Bob','Charlie','David']
})

orders_df = pd.DataFrame({
    'OrderID':[101,102,103],
    'CustomerID':[1,2,1],
    'OrderTotal':[250,150,200]
})

merged_df = pd.merge(clients_df, orders_df, left_on='ClientID', right_on='CustomerID',how='left')
print(merged_df)
```
