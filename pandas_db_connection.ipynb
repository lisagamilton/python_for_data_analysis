NULL VALUES

```python
import pandas as pd
data = {
    'Name':['John Doe','Jane Smith','Bob Brown','Alice Johnson'],
    'Age':[28,34, None, 45],
    'Email':['john@example.com',None,'bob@example.com','alice@example.com']
}

df = pd.DataFrame(data)
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Email'] = df['Email'].fillna('unknown@example.com')
df
```

```python
#WRONG
data = {
    'Name':['John Doe','Jane Smith','Bob Brown','Alice Johnson'],
    'Age':[28,34, None, 45],
    'Email':['john@example.com',None,'bob@example.com','alice@example.com']
}

df = pd.DataFrame(data)
df = df['Email'].dropna()
print(type(df))
```

```python
#CORRECT
data = {
    'Name':['John Doe','Jane Smith','Bob Brown','Alice Johnson'],
    'Age':[28,34, None, 45],
    'Email':['john@example.com',None,'bob@example.com','alice@example.com']
}

df = pd.DataFrame(data)

df = df.dropna(subset = 'Email')
print(type(df))
```

```python
from google.cloud import bigquery
from google.oauth2 import service_account
import pandas as pd
```

```python
credentials = service_account.Credentials.from_service_account_file('/content/credentials.json')
project_id = 'my-project-26-01-2026'
client = bigquery.Client(credentials = credentials, project=project_id)
query_text = f"""SELECT * FROM `my-project-26-01-2026.dataset1.subscriptions`;"""
query = client.query(query_text)
subscriptions = query.to_dataframe()
subscriptions
```

```python
subscriptions.info()
```

```python
#in this case it is better to delete duplicates (after asking the team/business decision maker)
subscriptions = subscriptions.drop_duplicates()
subscriptions.info()
```

```python
#It’s better to convert IDs to strings because they are identifiers, not values for calculation,
#and keeping them as strings preserves formatting and prevents data mismatches.
subscriptions['customer_id'] = subscriptions['customer_id'].astype('str')
subscriptions['subscriprion_id'] = subscriptions['subscriprion_id'].astype('str')
```

```python
##in this case it is better to delete Null values (after asking the team/business decision maker)
subscriptions = subscriptions.dropna()
subscriptions.describe()
```

```python
subscriptions = subscriptions.dropna()
subscriptions.info()
```

```python
print(subscriptions['plan'].unique())
```

```python
import re
text = 'Starter - 2022'
year = re.search(r'\d+',text).group()
print(year)
```

```python
subscriptions['plan_year'] = subscriptions['plan'].apply(lambda x: re.sub(r"[^0-9]+","",x))
subscriptions
```

```python
#Filtering out 'test' subscriptions
# ~ is 'NOT' operator - making everything opposite
subscriptions = subscriptions.loc[~subscriptions['plan'].str.contains('test')]
subscriptions.info()
```

```python
#we can change the empty rows like this
new_subscriptions = subscriptions.loc[subscriptions['plan_year'].str.contains(r'\d')]
new_subscriptions
```

```python
#or like this
subscriptions.loc[subscriptions['plan_year'] == "", 'plan_year'] = 2021
subscriptions
```

```python
subscriptions['plan_cleaned'] = subscriptions['plan'].str.extract(r'([A-Za-z]*)')
subscriptions
```

```python
#Creating new column 'plan_group', assigning 'Small team' or 'Big team'
for index, row in subscriptions.iterrows():
    if row['plan_cleaned'] in ('Starter', 'Squad'):
        subscriptions.loc[index, 'plan_group'] = 'Small team'
    else:
        subscriptions.loc[index, 'plan_group'] = 'Big team'
subscriptions
```

```python
sales_by_plan = subscriptions.groupby('plan_cleaned')['amount'].sum()
sales_by_plan
```

```python
df = df.dropna(subset = ['user_id','subscription_id','revenue','plan'])
print(df.info())
df
```

```python
#Converting to correct data type
df['user_id'] = df['user_id'].astype('str')
df['subscription_id'] = df['subscription_id'].astype('str')
df['signup_date'] = pd.to_datetime(df['signup_date'])
print(df.info())
df
```

```python
#revenue analysis
df['revenue_per_user'] = df.groupby('user_id')['revenue'].transform('sum')
df
```

```python
df['plan_group'] = df['plan_clean'].apply(lambda x: 'Small team' if x in ('Starter','Squad') else 'Big team')
df
```

```python
#What plan brings the most revenue?
df.groupby('plan_clean')['revenue'].sum().sort_values(ascending = False)
```

```python
#What is the (overall) churn rate?
canceled_users = len(df.loc[df['status'] == 'canceled'])
churn_rate = canceled_users/len(df)
print(f'Churn rate for the company is: {churn_rate:.2%}')
```

```python
#What is the (monthly) churn rate?
df['month'] = df['signup_date'].dt.to_period('M')
churn_rate = df.groupby('month')['status'].apply(lambda x: (x == 'canceled').sum()/len(x))
churn_rate
```

```python
#What is the churn rate regarding plan?
plan_churn_rate = df.groupby('plan_clean')['status'].apply(lambda x: (x == 'canceled').sum()/len(x))
plan_churn_rate
```

```python
#Який регіон найбільш активний?
df[df['status'] == 'active'].groupby('region').size().sort_values(ascending=False)
```
