```python
from google.cloud import bigquery
from google.oauth2 import service_account
import pandas as pd
```

```python
credentials = service_account.Credentials.from_service_account_file('/content/credentials.json')
```

```python
project_id = 'my-project-26-01-2026'
client = bigquery.Client(credentials = credentials, project=project_id)
query_text = f"""SELECT * FROM `my-project-26-01-2026.dataset1.customers_data`
LIMIT 100;"""
query = client.query(query_text)
```

```python
customers_data = query.to_dataframe()
customers_data
```

```python
customers_data.info()
```

```python
query_text1 = f"""SELECT * FROM `my-project-26-01-2026.dataset1.shopping_data`;"""
query1 = client.query(query_text1)
```

```python
shopping_data = query1.to_dataframe()
shopping_data
```

```python
shopping_data.info()
```

```python
shopping_data.describe()
```

```python
united_data = pd.merge(customers_data, shopping_data,how='left',on='customer_id')
united_data
```

```python
united_data.info()
```

```python
united_data['price'] = united_data['price'].replace(',','.',regex=True).astype('float')
united_data['category'] = united_data['category'].fillna('UNKNOWN')
united_data['invoice_date'] = pd.to_datetime(united_data['invoice_date'])
united_data.info()
```

```python
united_data.to_gbq(
    destination_table="dataset1.united_data",
    project_id="my-project-26-01-2026",
    credentials=credentials
)
```
