```python
import pandas as pd
df = pd.read_csv('/content/aug_train.csv')
```

# Pivot table

```python
def education_group(x):
  if x in ['High School','Primary School']:
    return 'School'
  if x in ['Masters','Phd']:
    return 'Masters and Phd'
  return x
```

```python
df['new_educational_level'] = df['education_level'].apply(education_group)
```

```python
df.groupby('new_educational_level')['target'].agg(['count','mean'])
```

```python
df.groupby('new_educational_level')['target'].agg(['count','mean'])
```

```python
df['new_educational_level'] = df['new_educational_level'].fillna('no_info')
```

```python
t = df.pivot_table(index='new_educational_level', values='target',aggfunc=['count','mean'])
```

```python
t.columns = ['_'.join(col).strip() for col in t.columns.values]
t
```

```python
t.reset_index()
```

Even though groupby is more comfortable in using in indexing, but pivot_table can add columns to the pivot!
It is useful when creating heatmaps

```python
df.pivot_table(index='new_educational_level',columns='gender', values='target',aggfunc=['count','mean'])
```
