# JOINS

```python
age = df[['enrollee_id']].copy()
```

```python
age['age'] = 30
age
```

```python
df = df.merge(age,how='left',on='enrollee_id')
df
```

# loc, iloc



```python
#show only females with city and target as columns
df.loc[df['gender'] == 'Female',['city','target']]
```

```python
#Select people with training_hours > 100 and education_level = Masters, and display all columns.
df.loc[(df['training_hours'] > 100) & (df['education_level'] == 'Masters'),:]
```

```python
#Modify: for all rows where training_hours > 100, set high_training = 1.
df.loc[df['training_hours'] > 100, 'high_training'] = 1
df
```

```python
#Replace: all NaN values in gender with No_info
df['gender'] = df['gender'].fillna('No_info')
df
```

```python
#Select: the first 5 rows and all columns
df.iloc[:5,:]
```

```python
#Select: the first 5 rows and only the first 3 columns
df.iloc[:5,:3]
```

```python
#Select: the 10th row and the 5th column
df.iloc[10,5]
```

```python
#last 3 rows
df.iloc[:,-3:]
```
