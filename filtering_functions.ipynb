```python
fruits = ["apple", "banana", "cherry", "banana"]
new_fruits = []
for f in fruits:
  if f == 'apple':
    new_fruits.append('orange')
  else:
    new_fruits.append(f)
print(new_fruits)
```

```python
#same but using list comprehension
[f if f != 'apple' else 'orange' for f in fruits]
```
