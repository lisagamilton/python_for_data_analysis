# What do you do with Nulls?
Nothing, we have no right to distort data, we just see the picture. When in data science building an algorithm, if there is 30% null values, it means we will not be able to process 30% people and business will lose money.
###However, when doing EDA (exploratory data analysis) we don't change anything, just see data.
