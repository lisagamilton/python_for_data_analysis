```python
# Three atempts
# Yes within 3 attempts - 'Glad we are on the same page'
# Otherwise - '3 strikes, you're out!'
attempts = 0
while attempts < 3:
  answer = input('Do you agree? yes/no?: ')
  if answer == 'yes':
    print('Glad we are on the same page!')
    break
  attempts +=1
else:
  print('''3 attempts, you're out!''')
```

```python
# how many numbers bigger than 3
numbers = [1,2,3,4,5,6]
count = 0
for n in numbers:
  if (n > 3):
    count+=1

print(count)
```

```python
#enumerate
#why use it? to find the exact position of the bad data in your list
letters = ['a','','c']
print(list(enumerate(letters,start = 1)))
for index, value in enumerate(letters):
  print(index,value)
#here we found the exact position of empty string
```

```python
letters = ['a','b','c']
numbers = [1,2,3]
for l, n in zip(letters,numbers):
  print(l,n)
```
